```python
# Dropping duplicate values
data2=data1.drop_duplicates(subset=["User_ID","Book_Title"])
data2
```

```python
data2[data2.duplicated()] 
```

```python
#number of unique users in the dataset
len(data2['User_ID'].unique())
```

```python
book=data2.pivot(index='User_ID',columns='Book_Title',values='Book_Rating').reset_index(drop=True)
book                            
```

```python
book.fillna(0, inplace=True)
book
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
book_sim=1-pairwise_distances(book.values,metric='cosine')
```

```python
book_sim
```

```python
#Storingresults in a dataframe
book_sim_df = pd.DataFrame(book_sim)
```

```python
#Set the index and column names to user ids 
book_sim_df.index=data2.User_ID.unique()
book_sim_df.columns=data2.User_ID.unique()
```

```python
book_sim_df
```

```python
# Nullifying diagonal values
np.fill_diagonal(book_sim, 0)
book_sim_df
```

```python
#Most Similar Users
book_sim_df.idxmax(axis=1)
```
